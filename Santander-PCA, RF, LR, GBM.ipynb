```python
from sklearn.decomposition import PCA
```

```python
X=df.drop(['target', 'ID_code'],axis=1)
Y=df['target']
```

```python
X
```

```python
Y
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
x=pca.transform(X)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x[:,0], x[:,1], hue=Y)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x[:,0],Y)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x[:,1],Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx, testx, trainy, testy = train_test_split(x,Y, train_size=0.8, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(trainx,trainy)
```

```python
pred_train=rf.predict(trainx)
```

```python
from sklearn.metrics import roc_auc_score , classification_report
```

```python
roc_auc_score(trainy, pred_train)
```

```python
pred_test = rf.predict(testx)
```

```python
roc_auc_score(testy,pred_test)
```

```python
rf.feature_importances_
```

```python
print(classification_report(testy,pred_test))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(trainx,trainy)
```

```python
y_pred = model.predict(testx)
```

```python
print(classification_report(testy,y_pred))
```
